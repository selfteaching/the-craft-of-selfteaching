## 质数计算流程优化
**为啥用了 $\sqrt{n}$ 就流程就优化了？**

以数字 100 为例，100可拆分为
- 2*50
- 4*25
- 5*20
- 10*10
- 20*5
- 25*4
- 20*2
- 其中后3个与前3个拆分的数字是一样的，所以验证前三个后就不用验证后3个了；理论上程序少跑了近一半的流程；

**为啥要 $\sqrt{n}$  +1 ？**

- $\sqrt{9}$ =3
- 那么`range(1,3)`的取值范围是不包含 3 的， 9 就会被错误判定为质数；


```python
import random
r = random.randrange(2,13)

if r == 7:
    print('平')
elif r < 7:
    print('小')
elif r > 7:
    print('大')
```

```python
import random
r = random.randrange(0,13)

if r < 2:
    print('投飞了')
elif r >= 2 and r < 7:
    print('小')
elif r == 7:
    print('平')
elif r > 7:
    print('大')
```

```python
for n in range(2,100):
    if n == 2:
        print(n)
        continue
    for i in range(2,int(n**0.5)+1):
        if n % i == 0:
            break
    else:
        print(n)
```

```python
def is_prime(n):
    if n==2:
        return True
    for m in range(2,int(n**0.5)+1):
        if n%m==0:
            return False
    else:
        return True

for i in range(100,200):
    if is_prime(i):
        print(i)
```
